```python
n = int(input())

def fib(n):
    if n == 1 or n == 2:
        return 1  # 1
    else:
        return (fib(n - 1) + fib(n - 2))
    
def fibonacci(n):
    num=0
    f = [0]*(n+1)
    f[2] = 1
    f[1] = f[2]
    for i in range(3, n+1):
        num += 1
        f[i] =  f[i - 1] + f[i - 2]
    return num

print(fib(n), fibonacci(n))
```
